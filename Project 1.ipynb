## 1A: English Dictionary App

### This project uses the skills learnt from week 1 to week 6. The aim of this project is to create an English Dictionary app that returns definitions of English words. When the application is started, it displays a menu as follows:

#### Main Menu

#### Add a new word
#### Find the meaning
#### Update a word
#### Exit

#### Enter Choice:

#### If item 1 is selected, it should prompt for a word and then accept its meaning. This pair of word and its meaning should be stored in a file called ‘words.txt’. When item 2 is selected, it should prompt for a word, search that word in the file ‘words.txt’ and return the meaning if it is found. If the meaning is not found, it should display an appropriate message. Similarly, when choice 3 is entered, it should accept a word and update its meaning. In all the above three cases, the prompt should return to the main menu. The words and their meanings should be stored as a dictionary in words.txt, using the concept of serialization. If choice 4 is selected, a graceful exit should be performed.

```python
import json
import os
```

```python
fileName = "words.txt" 

if os.path.isfile(fileName):
    f = open(fileName,'a')
    print(' File already Exists \n')   
else:
    f = open(fileName,'w') 
    dic = {'add':'sum'}
    json.dump(dic,f)
    
f.close()  

def add_words(word,meaning):  
    
    f1 = open("words.txt",'r')  
    
    temp_dict = json.load(f1)
    
    f1.close 
    
    if word in temp_dict.keys():
        print("_"*100)
        print ("%s already exists if you want to update the meaning of %s use option 3" % (word,word))
        print("_"*100)
    else:
        f2 = open("words.txt",'w')  
        temp_dict[word] = meaning
        json.dump(temp_dict,f2)
        f2.close                      
    

def get_meaning(word):
    
    f3 = open("words.txt",'r')
    
    temp_dict = json.load(f3)
    
    f3.close
    
    if word in temp_dict.keys():
        print("_"*100)
        print ('meaning of %s is : %s' % (word,temp_dict[word]))
        print("_"*100)
    else:
        print("_"*100)
        print('meaning of %s not found choose option 3 to enter its meaning' %(word))
        print("_"*100)
        

def update_word(word):  
    
    f4 = open("words.txt",'r')
    
    temp_dict = json.load(f4)
    
    f4.close
    
    if word in temp_dict.keys():
        print ("Enter the new meaning :")       
        new_meaning = input("-> ")
        print ("_"*100)
        f5 = open("words.txt",'w')
        temp_dict[word] = new_meaning
        json.dump(temp_dict,f5)
        f5.close  
    else:
        print("_"*100)
        print ('No word such as %s found in the dictionay to add its meaning use option 1\n' % (word))
        print("_"*100)

     
while True:
    print ("ENGLISH Dictionary APP ")
    print ("_"*100)           
    print ("Enter 1 to add a new word to the Dictionary")
    print ("Enter 2 to find the meaning of the word in the Dictionary")
    print ("Enter 3 to update the new meaning of a word")
    print ("Enter 4 to EXIT the Dictionary")
    print ("_"*100)
    choice = input('Choose any option from the given choices : ')
    
    if choice == "1":
        print ('Enter the word for which you want to add meaning for :')       
        word1 = input("-> ")
        print ('Enter the meaning of the above word :') 
        meaning1 = input("-> ")
        print ("_"*100)
        add_words(word1,meaning1)
        
    elif choice == "2":
        print ('Enter the word for whcich you want to know the meaning :')       
        word2 = input("-> ")
        get_meaning(word2)
        
    elif choice == "3":
        print ('Enter the word for which you want to update meaning for :')       
        word3 = input("-> ")
        update_word(word3)
        
    elif choice == "4":
        print("_"*100)
        print("THANK YOU FOR USING THE Dictionary")
        print("_"*100)
        break   
```

## 1B: Library Book Management System

### In this project, a Library Book Management System will be implemented with the front end GUI developed using Tkinter and the backend database implemented in SQLite. It is a single window tkinter GUI application which will implement CRUD( create, read, update, delete) operations on a ‘books’ database, in SQLite. The window will contain text widgets to accept / display book title, author, year of publication, ISBN (optional) and a large text area/ list box to display the records which match the search criteria.

```python
from tkinter import *
import tkinter.ttk as ttk
import tkinter.messagebox as tkMessageBox
import sqlite3
```

```python
root=Tk()
root.title("....Libary Management Application....")

#------------------------------------------Fuctions-----------------------------------------------------------
def Database():
    global conn,cursor
    conn = sqlite3.connect('books.db')
    cursor = conn.cursor()
    cursor.execute("CREATE TABLE IF NOT EXISTS lab(booktitle TEXT,author TEXT,year TEXT,isbn TEXT)")

def Add():
        Database()
        if BOOKTITLE.get()=="" or AUTHOR.get()=="" or YEAR.get()==""or ISBN.get()=="":
            txt_result.config(text="Please enter all values",fg="red")
        else:
            cursor.execute("INSERT INTO lab(booktitle,author,year,isbn) values(?,?,?,?)",[str(BOOKTITLE.get()),str(AUTHOR.get()),str(YEAR.get()),str(ISBN.get())])
            conn.commit()
            BOOKTITLE.set("")
            AUTHOR.set("")
            YEAR.set("")
            ISBN.set("")
            cursor.close()
            conn.close()
            txt_result.config(text="Record Inserted",fg="green")

def Display():
    Database()
    cursor.execute("SELECT * FROM lab ORDER BY booktitle ASC")
    fetch = cursor.fetchall()
    for data in fetch:
        tree.insert('','end',values=(data[0],data[1],data[2],data[3]))
    cursor.close()
    txt_result.config(text="Successful fetch data",fg="black")
    
def Search():
    Database()
    cursor.execute("SELECT * FROM lab WHERE booktitle = ? or author = ? or year = ? or isbn=? ",[str(BOOKTITLE.get()),str(AUTHOR.get()),str(YEAR.get()),str(ISBN.get())])
    rows=cursor.fetchall()
    for data in rows:
        tree.insert('','end',values=(data[0],data[1],data[2],data[3]))
    cursor.close()

def Issue():
    Database()
    txt_result.config(text="Book is Issue.Thank you",fg = "green")
        
def Delete():
    Database()
    cursor.execute("DELETE FROM lab where booktitle = ? or author = ? ",(BOOKTITLE.get(),AUTHOR.get(),))
    conn.commit()
    ISBN.set("")
    BOOKTITLE.set("")
    AUTHOR.set("")
    YEAR.set("")
    cursor.close()
    conn.close()
    #cursor.close()
    txt_result.config(text="Successfully Deleted ")
    
def Exit():
    result = tkMessageBox.askquestion('Do you want to exit? (y/n)',icon='warning')
    if result == "yes":
        root.destroy()
        exit() 
#-----------------------------------------------Variable Decleartion-------------------------------------------------

BOOKTITLE = StringVar()
AUTHOR = StringVar()
YEAR = StringVar()
ISBN = StringVar()

#--------------------------------------------------Frame------------------------------------------------------------- 

Top = Frame(root,width = 900,height = 50,bd=8,relief = 'raise')
Top.pack(side = TOP)

Left = Frame(root,width = 300,height = 500,bd=8,relief = 'raise',bg = "red")
Left.pack(side = LEFT)

Right = Frame(root,width = 600,height = 500,bd=8,relief = 'raise',bg = "yellow")
Right.pack(side = RIGHT)

Forms = Frame(Left,width = 300,height = 450)
Forms.pack(side = TOP)

Buttons = Frame(Left,width = 300,height = 100,relief = 'raise')
Buttons.pack(side = BOTTOM)
# -------------------------------------------------Label-------------------------------------------------------------
txt_title = Label(Top,width=900,font=('arial',24),text='Libary Book Management')
txt_title.pack()

txt_booktitle = Label(Forms,font=('arial',16),text='Book Title',bd=15)
txt_booktitle.grid(row=0,stick='e')

txt_author = Label(Forms,font=('arial',16),text='Author',bd=15)
txt_author.grid(row=1,stick='e')

txt_year = Label(Forms,font=('arial',16),text='Year',bd=15)
txt_year.grid(row=2,stick='e')

txt_isbm = Label(Forms,font=('arial',16),text='ISBN',bd=15)
txt_isbm.grid(row=3,stick='e')

txt_result = Label(Buttons)
txt_result.pack(side=TOP)
#--------------------------------------------------Entry------------------------------------------------------------
booktitle = Entry(Forms,textvariable = BOOKTITLE,width=30)
booktitle.grid(row=0,column=1)

author = Entry(Forms,textvariable = AUTHOR,width=30)
author.grid(row=1,column=1)

year = Entry(Forms,textvariable = YEAR,width=30)
year.grid(row=2,column=1)

isbm = Entry(Forms,textvariable = ISBN,width=30)
isbm.grid(row=3,column=1)

#---------------------------------------------Buttons----------------------------------------------------------
but_read = Button(Buttons,width = 10,text = 'Display',command=Display)
but_read.pack(side=LEFT)

but_search = Button(Buttons,width = 10,text = 'Search',command=Search)
but_search.pack(side=LEFT)

but_create = Button(Buttons,width = 10,text = 'Add',command=Add)
but_create.pack(side=LEFT)

but_issue = Button(Buttons,width = 10,text = 'Issue',command=Issue)
but_issue.pack(side=LEFT)

but_delete = Button(Buttons,width = 10,text = 'Delete',command=Delete)
but_delete.pack(side=LEFT)

but_exit = Button(Buttons,width = 10,text = 'Exit',command=Exit)
but_exit.pack(side=LEFT)

#----------------------------------------------------List-------------------------------------------------------
scrollbary = Scrollbar(Right,orient=VERTICAL)
scrollbarx = Scrollbar(Right,orient=HORIZONTAL)

tree = ttk.Treeview(Right,columns=("BookTitle","Author","Year","Isbn"))

scrollbary.config(command=tree.yview)
scrollbary.pack(side=RIGHT,fill=Y)

scrollbarx.config(command=tree.xview)
scrollbary.pack(side=BOTTOM,fill=X)

tree.heading('BookTitle',text="Booktitle",anchor=W)
tree.heading('Author',text="Author",anchor=W)
tree.heading('Year',text="Year",anchor=W)
tree.heading('Isbn',text="ISBN",anchor=W)

tree.column("#0",stretch=NO,minwidth=0,width=20)
tree.column("#1",stretch=NO,minwidth=0,width=180) 
tree.column("#2",stretch=NO,minwidth=0,width=60)
tree.column("#3",stretch=NO,minwidth=0,width=60)
tree.column("#4",stretch=NO,minwidth=0,width=80)
tree.pack()
#------------------------------------------Ending loop---------------------------------------------------------------
if __name__ == "__main__":
    root.mainloop()
```
